# Multi Class Classification with SVM

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm
from sklearn.datasets import make_blobs

import warnings
warnings.filterwarnings('ignore')

%matplotlib inline
```

## Create 5 classes of data

```python
# Create 40 separable points:
X, y = make_blobs(n_samples = 200, centers = 5, random_state = 0)
```

### Define the functions for plotting

```python

def make_meshgrid(x, y, h=.02):
    """Create a mesh of points to plot in

    Parameters
    ----------
    x: data to base x-axis meshgrid on
    y: data to base y-axis meshgrid on
    h: stepsize for meshgrid, optional

    Returns
    -------
    xx, yy : ndarray
    """
    x_min, x_max = x.min() - 1, x.max() + 1
    y_min, y_max = y.min() - 1, y.max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))
    return xx, yy
```

```python

def plot_contours(ax, clf, xx, yy, **params):
    """Plot the decision boundaries for a classifier.

    Parameters
    ----------
    ax: matplotlib axes object
    clf: a classifier
    xx: meshgrid ndarray
    yy: meshgrid ndarray
    params: dictionary of params to pass to contourf, optional
    """
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    out = ax.contourf(xx, yy, Z, **params)
    return out

```

## Plot Data

```python
plt.scatter(X[:, 0], X[:, 1], c = y, s = 30, cmap = plt.cm.Paired)
plt.show()
```

## Generate SVM Models

```python

cmap = plt.cm.Paired
C = 1.0  # SVM regularization parameter
models = (svm.SVC(kernel='linear', C=C,decision_function_shape='ovo'),
          svm.LinearSVC(C=C, max_iter=10000),
          svm.SVC(kernel='rbf', gamma=0.7, C=C,decision_function_shape='ovo'),
          svm.SVC(kernel='poly', degree=3, gamma='auto', C=C,decision_function_shape='ovo'))

# Fit all the models
models = (clf.fit(X, y) for clf in models)

# title for the plots
titles = ('SVC with linear kernel',
          'LinearSVC (linear kernel)',
          'SVC with RBF kernel',
          'SVC with polynomial (degree 3) kernel')

# Plot the decision boundaries for each model
fig, sub = plt.subplots(2, 2, figsize=(15,15))
plt.subplots_adjust(wspace=0.2, hspace=0.2)

X0, X1 = X[:, 0], X[:, 1]
xx, yy = make_meshgrid(X0, X1)

for clf, title, ax in zip(models, titles, sub.flatten()):
    plot_contours(ax, clf, xx, yy,
                  cmap=cmap, alpha=0.8)
    ax.scatter(X0, X1, c=y, cmap=cmap, s=30)
    ax.set_xlim(xx.min(), xx.max())
    ax.set_ylim(yy.min(), yy.max())
    ax.set_xticks(())
    ax.set_yticks(())
    ax.set_title(title+': '+str(clf.score(X, y)))

plt.show()
```

```python

def optimize_nuSVC(X,y):
    best = {'nu': 0, 'kernel': 'none', 'degree': 3, 'gamma': 'auto', 'score': 0}

    nu = [0.2,0.3,0.4,0.5]
    kernel = ['linear', 'poly', 'rbf', 'sigmoid']
    degree = [3,4,5]
    gamma = ['auto',0.2,0.4,0.5,0.6,0.8,0.9]

    for n in nu:
        for k in kernel:
            for d in degree:
                for g in gamma:
                    clf = svm.NuSVC(nu=n, kernel=k, degree=d, gamma=g)
                    clf.fit(X, y)
                    score = clf.score(X, y)
                    
                    if score > best['score']:
                        best = {'nu': n, 
                                'kernel': k, 
                                'degree': d, 
                                'gamma': g, 
                                'score': score}

    return best
```

```python
optimum = optimize_nuSVC(X,y)
print (optimum)
```

```python
title = 'NuSCV with '+optimum['kernel']+ ' kernel (degree: '+ \
        str(optimum['degree'])+', Gamma: '+str(optimum['gamma'])+')'

clf = svm.NuSVC(nu=optimum['nu'], 
                kernel=optimum['kernel'],
                degree=optimum['degree'], 
                gamma=optimum['gamma'])

clf.fit(X, y)

# Plot the decision boundaries for each model
fig = plt.figure(figsize=(15,15))
ax = plt.gca()
X0, X1 = X[:, 0], X[:, 1]
xx, yy = make_meshgrid(X0, X1)


ax.scatter(X0, X1, c=y, cmap=plt.cm.Paired, s=50)
ax.set_xlim(xx.min(), xx.max())
ax.set_ylim(yy.min(), yy.max())
ax.set_xticks(())
ax.set_yticks(())
ax.set_title(title+': \nScore: '+str(clf.score(X, y)))
   
plot_contours(ax, clf, xx, yy, cmap=cmap, alpha=0.5)
    

plt.show()
```
