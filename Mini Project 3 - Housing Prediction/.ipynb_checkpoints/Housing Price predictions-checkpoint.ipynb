## Some supporting functions

```python
def missing_bar(df,figsize=(20,20)):
    total = df.shape[0]
    missing = (total - df.isnull().sum())/total * 100
    missing.plot(kind='barh',figsize=figsize)
    plt.gca().invert_yaxis()
    plt.xlabel('Percentage of Available Data')
    plt.ylabel('Feature Label')
    plt.title('Valid Values', fontsize=14)
    plt.show()

    print('{}% of data is missing '.format(round(df.isnull().sum().sum()/float(df.size)*100,3)))
    
```

```python
def average_ralation_to_feature(df, groupby, col, precision=3):
    '''
    Find the average value for each group which can be applied to the new column
    
    df: The dataframe.
    
    groupby: The column that has the relation to the feature with missing data.
    
    col: The feature with the missing data.
    
    precision: The precision of the results
    (Default: 3)
    '''
    
    # calcualte the average value for each col
    groups = df.groupby(by=groupby)[col].mean()
    groups = round(groups,precision)
    groups[groups.isnull() == True] = df[col].mean()

    # Populate the missing data with the average value for the corresponding grouped column.
    for item in groups.index:
        if (item in groups.keys()):
            df[col][(df[col].isnull() == True) & (df[groupby] == item)] = groups[item]

    # Replace any remaining NaN values with the feature average.
    df[col][df[col].isnull() == True] = df[col].mean()
    
    
    return df
```

```python
# Create some supporting plotting functions
def plot_scatter_grid(data, possible_predictors,col=4):
    
    row = int(len(possible_predictors)/col)+1
    
    
    target = data['SalePrice']
    fig = plt.figure(figsize=(20,row*4))
    fig.subplots_adjust(hspace=0.4, wspace=0.4)
    for idx, predictor in enumerate(possible_predictors):
        ax = fig.add_subplot(row, col, idx+1)
        plt.scatter(data[predictor], target, s=1)
        ax.set_title(predictor)
        
def plot_distribution_grid(data, possible_predictors,col=4):
    
    row = int(len(possible_predictors)/col)+1
    
    target = data['SalePrice']
    fig = plt.figure(figsize=(20,row*4))
    fig.subplots_adjust(hspace=0.4, wspace=0.4)
    for idx, predictor in enumerate(possible_predictors):
        plot_data = data[predictor][data[predictor].isnull() == False]
        ax = fig.add_subplot(row, col, idx+1)
        sns.distplot(plot_data)
        ax.set_title(predictor)
        
def plot_box_grid(data, possible_predictors,col=4):
    
    row = int(len(possible_predictors)/col)+1
    
    target = data['SalePrice']
    fig = plt.figure(figsize=(20,row*4))
    fig.subplots_adjust(hspace=0.4, wspace=0.4)
    for idx, predictor in enumerate(possible_predictors):
        ax = fig.add_subplot(row, col, idx+1)
        sns.boxplot(x=data[predictor])
        ax.set_title(predictor)


def plot_correlation_matrix(data,figsize=(15,15)):
    corr_mat=data.corr(method='pearson')
    plt.figure(figsize=figsize)
    sns.heatmap(corr_mat,vmax=1,square=True,annot=True,cmap='cubehelix')
    
```

## Feature Engineering
A reasonable predictor of the price of the house is the average price in the area and by building class in the previous year.

```python
# Last years average price
averageSalePrice = housing.groupby(by=['YrSold'])['SalePrice'].mean()

housing['AveragePrice_LastYear'] = averageSalePrice[2006]
for year in averageSalePrice[1:].index:
    housing['AveragePrice_LastYear'][housing['YrSold'] == year] = averageSalePrice[year-1]

```

```python

# Last years average neighborhood price
averageNeighborhoodPrice = housing.groupby(by=['YrSold','Neighborhood'])['SalePrice'].mean()
averageNeighborhoodPrice = averageNeighborhoodPrice.unstack()
for year in averageNeighborhoodPrice.index:
    averageNeighborhoodPrice.loc[year][averageNeighborhoodPrice.loc[year].isnull() == True] = \
    averageNeighborhoodPrice.mean(axis=1)[year]

housing['AverageNeighborhoodPrice_LastYear'] = averageNeighborhoodPrice.mean(axis=1)[2006]
for year in averageNeighborhoodPrice.index[1:]:
    for neighborhood in averageNeighborhoodPrice.columns:
        mask = (housing['YrSold'] == year) & (housing['Neighborhood'] == neighborhood)
        housing['AverageNeighborhoodPrice_LastYear'][mask] = averageNeighborhoodPrice.loc[year-1][neighborhood]
        
        
        
```

```python
# Last years average price by building class
averageClassPrice = housing.groupby(by=['YrSold','MSSubClass'])['SalePrice'].mean()
averageClassPrice = averageClassPrice.unstack()
for year in averageClassPrice.index:
    averageClassPrice.loc[year][averageClassPrice.loc[year].isnull() == True] = \
    averageClassPrice.mean(axis=1)[year]

housing['AverageClassPrice_LastYear'] = averageClassPrice.mean(axis=1)[2006]
for year in averageClassPrice.index[1:]:
    for neighborhood in averageClassPrice.columns:
        mask = (housing['YrSold'] == year) & (housing['MSSubClass'] == neighborhood)
        housing['AverageClassPrice_LastYear'][mask] = averageClassPrice.loc[year-1][neighborhood]
        
        
              
```

### Outlier Detection

```python
def split_data(data, features, target='SalePrice', split_col='Yrsold', split_val=2010):
    # Split the data into training and testing
    

    training = data[data[split_col] < split_val]
    testing = data[data[split_col] >= split_val]

    # Seperate the trainig and testing features
    X_train = training[features]
    X_test = testing[features]
    y_train = training[target]
    y_test = testing[target]
    
    return X_train, X_test, y_train, y_test


def scale_data(X_train, X_test, y_train, y_test, features):
    
    scalerX = StandardScaler().fit(X_train)
    scalery = StandardScaler().fit(y_train.values.reshape(-1, 1))
    
    X_train = scalerX.transform(X_train)
    y_train = scalery.transform(y_train.values.reshape(-1, 1))
    X_test = scalerX.transform(X_test)
    y_test = scalery.transform(y_test.values.reshape(-1, 1))

    # Create dataframes for the training and testing data
    X_train = pd.DataFrame(X_train,columns=features)
    X_test = pd.DataFrame(X_test,columns=features)
    y_train = pd.DataFrame(y_train,columns=['SalePrice'])
    y_test = pd.DataFrame(y_test,columns=['SalePrice'])
    
    return X_train, X_test, y_train, y_test, scalerX, scalery
    
```

### Forward Feature Selection

```python
features = list(set(list(housing.columns)) - set(['SalePrice']))

X_train, X_test, y_train, y_test = \
split_data(housing, features, target='SalePrice', split_col='YrSold', split_val=2010)

X_train_sc, X_test_sc, y_train_sc, y_test_sc, scalerX, scalery = \
scale_data(X_train, X_test, y_train, y_test, features)


```

```python
## Use Forward Feature Selection to pick a good model

# start with no predictors
included = []
# keep track of model and parameters
best = {'feature': '', 'r2': 0, 'a_r2': 0}
# create a model object to hold the modelling parameters
model = RidgeCV(cv=5) #BaggingRegressor()
# get the number of cases in the test data
n = X_test.shape[0]

r2_list = []
adjusted_r2_list = []
```

```python
while True:
    changed = False
    
    # list the features to be evaluated
    excluded = list(set(X_test.columns) - set(included))
        
    # for each remaining feature to be evaluated
    for new_column in excluded:
        
        # fit the model with the Training data
        fit = model.fit(X_train[included + [new_column]], y_train)
        # calculate the score (R^2 for Regression)
        score = model.score(X_test[included + [new_column]], y_test)
        r2_list.append(score)
        # number of predictors in this model
        k = len(included + [new_column]) + 1
        # calculate the adjusted R^2
        adjusted_r2_list.append(adjusted_r2(r2_list[-1],n,k))

        # if model improves
        if adjusted_r2_list[-1] > best['a_r2']:
            
            # record new parameters
            best = {'feature': new_column, 'r2': r2_list[-1], 'a_r2': adjusted_r2_list[-1]}
            # flag that found a better model
            changed = True
        

    # if found a better model after testing all remaining features
    if changed:
        # update control details
        included.append(best['feature'])
        excluded = list(set(excluded) - set(best['feature']))
        print('Added feature %-4s with R^2 = %.3f and adjusted R^2 = %.3f' % 
              (best['feature'], best['r2'], best['a_r2']))
    else:
        # terminate if no better model
        break
```

```python
## Use Forward Feature Selection to pick a good model
def feed_forward(model, X_train, y_train, X_test, y_test):

    # start with no predictors
    included = []
    # keep track of model and parameters
    best = {'feature': '', 'r2': 0, 'a_r2': 0}
    # create a model object to hold the modelling parameters
    #model = BaggingRegressor()
    # get the number of cases in the test data
    n = X_test.shape[0]

    r2_list = []
    adjusted_r2_list = []

    while True:
        changed = False

        # list the features to be evaluated
        excluded = list(set(X_test.columns) - set(included))

        # for each remaining feature to be evaluated
        for new_column in excluded:

            # fit the model with the Training data
            fit = model.fit(X_train[included + [new_column]], y_train)
            # calculate the score (R^2 for Regression)
            score = model.score(X_test[included + [new_column]], y_test)
            r2_list.append(score)
            # number of predictors in this model
            k = len(included + [new_column]) + 1
            # calculate the adjusted R^2
            adjusted_r2_list.append(adjusted_r2(r2_list[-1],n,k))

            # if model improves
            if adjusted_r2_list[-1] > best['a_r2']:

                # record new parameters
                best = {'feature': new_column, 'r2': r2_list[-1], 'a_r2': adjusted_r2_list[-1]}
                # flag that found a better model
                changed = True


        # if found a better model after testing all remaining features
        if changed:
            # update control details
            included.append(best['feature'])
            excluded = list(set(excluded) - set(best['feature']))
        else:
            # terminate if no better model
            break
            
    return included, best['a_r2']
```

```python
models = []

models.append(("LassoCV",LassoCV(cv=5)))
models.append(("LinearRegression",LinearRegression()))
models.append(("PassiveAggressiveRegressor",PassiveAggressiveRegressor()))
models.append(("RidgeCV",RidgeCV(cv=5)))
models.append(("Bagging",BaggingRegressor()))
models.append(("SVR",SVR()))

results = []
names = []
feature_list = {}

best_model = {'model': '', 'features': [], 'a_r2': 0}

best = 0
for name, model in models:
    print (name)
    features, result = feed_forward(model, X_train, y_train, X_test, y_test)
    
    if (result > best_model['a_r2']):
        best_model['model'] = model
        best_model['features'] = features
        best_model['a_r2'] = result
        
        
print (best_model)

    
    
```

Performing feed forward feature selection on all the models, results in the LinearRegression model being the best model to use, with an $R^{2}$ score of 0.913.

We can try to improve this by tunning our hyper parameters.

```python

def optimize_Bagging(X_train, y_train, X_test, y_test):
    max_features = [0.5, 0.6, 0.7, 0.8, 0.9, 1.0]
    n_estimators = [10,20,30,50,100,200]
    bootstrap = [False, True]
    n, k = X_test.shape
    
    optimum = {'max_features': 0.1, 'n_estimators': 1, 'bootstrap': '', 'score': 0}
    
    for f in max_features:
        for n_e in n_estimators:
            for boot in bootstrap:
                model = BaggingRegressor(n_estimators=n_e,max_features=f,bootstrap=boot)

                model.fit(X_train, y_train)

                y_pred = model.predict(X_test)

                score = model.score(X_test, y_test)
                adj_r2 = adjusted_r2(score,n,k)

                if adj_r2 > optimum['score']:
                    optimum = {'max_features': f,
                               'n_estimators': n_e,
                               'bootstrap': boot,  
                               'score': adj_r2}

    
    return optimum
                

def optimize_LinearRegression(X_train, y_train, X_test, y_test):
    fit_intercept = [False, True]
    normalise = [True, False]
    
    n, k = X_test.shape
    
    optimum = {'fit_intercept': True, 'normalise': False,  'score': 0}
    
    for intercept in fit_intercept:
        for norm in normalise:
            model = LinearRegression(fit_intercept=intercept,normalize=norm)

            model.fit(X_train, y_train)

            y_pred = model.predict(X_test)

            score = model.score(X_test, y_test)
            adj_r2 = adjusted_r2(score,n,k)

            if adj_r2 > optimum['score']:
                optimum = {'fit_intercept': intercept,
                           'normalise': norm,
                           'score': adj_r2}

    
    return optimum


def optimize_RidgeCV(X_train, y_train, X_test, y_test):
    alphas = (0.1, 0.5 ,1.0, 5.0, 10.0)
    fit_intercept = [False, True]
    normalise = [True, False]
    cross_validation = [3,5,7,10]
    gcv_mode = ['auto','svd','eigen']
    
    n, k = X_test.shape
    
    optimum = {'fit_intercept': True, 'normalise': False, 'cv': 1 ,'gcv_mode': 'auto', 'score': 0}
    
    
    for intercept in fit_intercept:
        for norm in normalise:
            for cv in cross_validation:
                for gcv in gcv_mode:
            
                    model = RidgeCV(fit_intercept=intercept, normalize=norm, cv=cv, gcv_mode=gcv)

                    model.fit(X_train, y_train)

                    y_pred = model.predict(X_test)

                    score = model.score(X_test, y_test)
                    adj_r2 = adjusted_r2(score,n,k)

                    if adj_r2 > optimum['score']:
                        optimum = {'fit_intercept': intercept,
                                   'normalise': norm,
                                   'cv': cv ,
                                   'gcv_mode': gcv,
                                   'score': adj_r2}

    
    return optimum
```
