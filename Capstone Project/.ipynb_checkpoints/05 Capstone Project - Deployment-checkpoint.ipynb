```python
from IPython.html import widgets
from IPython.display import display, Markdown
    
initialise = [[0]*len(included)]
dfp = pd.DataFrame(initialise,columns=included)
    
def reset_feature(feature):
    feature_cols = [col for col in dfp.columns if feature in col]
    dfp[feature_cols] = 0
    
    
            
            
def Select_Model(Model):
    reset_feature('Model')
    
    key = 'Model_'+Model
    if key in dfp.columns:
        dfp[key] = 1
        
def Select_Brand(Brand):
    reset_feature('Brand')
    
    model.options = Model_List[Brand]
    key = 'Brand_'+Brand
    if key in dfp.columns:
        dfp[key] = 1

def Select_State(State):
    reset_feature('State')
    
    key = 'State_'+State
    if key in dfp.columns:
        dfp[key] = 1

def Select_Traction(Traction_Control):
    reset_feature('Traction_Control')
    
    key = 'Traction_Control_'+Traction_Control
    if key in dfp.columns:
        dfp[key] = 1

def Select_Cooling(Cooling):
    reset_feature('Cooling')
    
    key = 'Cooling_'+Cooling
    if key in dfp.columns:
        dfp[key] = 1

def Select_Swingarm_Type(Swingarm_Type):
    reset_feature('Swingarm_Type')
    
    key = 'Swingarm_Type_'+Swingarm_Type
    if key in dfp.columns:
        dfp[key] = 1

def Select_Colour(Colour):
    reset_feature('Colour')
    
    key = 'Colour_'+Colour
    if key in dfp.columns:
        dfp[key] = 1

def Select_Exhaust_Config(Exhaust_Config):
    reset_feature('Exhaust_Config')
    
    key = 'Exhaust_Config_'+Exhaust_Config
    if key in dfp.columns:
        dfp[key] = 1

def Select_Front_Brake_Description(Front_Brake_Description):
    reset_feature('Front_Brake_Description')
    
    key = 'Front_Brake_Description_'+Front_Brake_Description
    if key in dfp.columns:
        dfp[key] = 1

def Select_Front_Brake_Caliper(Front_Brake_Caliper):
    reset_feature('Front_Brake_Caliper')
    
    key = 'Front_Brake_Caliper_'+Front_Brake_Caliper
    if key in dfp.columns:
        dfp[key] = 1

def Select_Rear_Brake_Description(Rear_Brake_Description):
    reset_feature('Rear_Brake_Description')
    
    key = 'Rear_Brake_Description_'+Rear_Brake_Description
    if key in dfp.columns:
        dfp[key] = 1

def Select_Rear_Brake_Caliper(Rear_Brake_Caliper):
    reset_feature('Rear_Brake_Caliper')
    
    key = 'Rear_Brake_Caliper_'+Rear_Brake_Caliper
    if key in dfp.columns:
        dfp[key] = 1

def Select_Frame_Material(Frame_Material):
    reset_feature('Frame_Material')
    
    key = 'Frame_Material_'+Frame_Material
    if key in dfp.columns:
        dfp[key] = 1

def Select_Carburettor(Carburettor):
    reset_feature('Carburettor')
    
    key = 'Carburettor_'+Carburettor
    if key in dfp.columns:
        dfp[key] = 1

def Select_Front_Suspension(Front_Suspension):
    reset_feature('Front_Suspension')
    
    key = 'Front_Suspension_'+Front_Suspension
    if key in dfp.columns:
        dfp[key] = 1

def Select_Side_Stand(Side_Stand):
    reset_feature('Side_Stand')
    
    key = 'Side_Stand_'+Side_Stand
    if key in dfp.columns:
        dfp[key] = 1

def Select_Cam_Type(Cam_Type):
    reset_feature('Cam_Type')
    
    key = 'Cam_Type_'+Cam_Type
    if key in dfp.columns:
        dfp[key] = 1

              
            
            
def predict_price(_):
    # The model needs to be provided.
    
    
    #clear_output()
    
#     print ('Model: ',model.value)
#     print ('Brand: ',brand.value)
#     print ('State: ',state.value)
#     print ('Traction_Control: ',traction.value)
#     print ('Engine_Size: ',engine.value)
#     print ('Age: ',age.value)
#     print ('Kilometers: ',kilometers.value)
#     print ('Registration: ',registration.value)
#     print ('Length: ',length.value)
#     print ('Trail: ',trail.value)
#     print ('Front_Brake_Diameter: ',front_brake.value)

    #predicted_price = model.predict(dfp)
    predicted_price = 1000000
    
    display(Markdown('# ${:,}'.format(predicted_price)))

    #print (dfp)
    
    
    




```

```python
from ipywidgets import Layout

style = {'description_width': 'initial',
         'width': '100px',
         'height': '100px'}


# Dropdown list
brand = widgets.Dropdown(options=Model_List.keys(), style=style)
init = brand.value
model = widgets.Dropdown(options=Model_List[init], style=style)
state = widgets.Dropdown(options=State_List, style=style)
traction = widgets.Dropdown(options=Traction_Control_List, style=style)
cooling = widgets.Dropdown(options=Cooling_List, style=style)
swingarm = widgets.Dropdown(options=Swingarm_Type_List, style=style)
colour = widgets.Dropdown(options=Colour_List, style=style)
exhaust = widgets.Dropdown(options=Exhaust_Config_List, style=style)
front_brake = widgets.Dropdown(options=Front_Brake_Description_List, style=style)
front_brake_caliper = widgets.Dropdown(options=Front_Brake_Caliper_List, style=style)
rear_brake = widgets.Dropdown(options=Rear_Brake_Description_List, style=style)
rear_brake_caliper = widgets.Dropdown(options=Rear_Brake_Caliper_List, style=style)
frame = widgets.Dropdown(options=Frame_Material_List, style=style)
carburettor = widgets.Dropdown(options=Carburettor_List, style=style)
front_suspension = widgets.Dropdown(options=Front_Suspension_List, style=style)
side_stand = widgets.Dropdown(options=Side_Stand_List, style=style)
cam_type = widgets.Dropdown(options=Cam_Type_List, style=style)

# Interactive widgets
model_DD = widgets.interactive(Select_Model, Model=model)
brand_DD = widgets.interactive(Select_Brand, Brand=brand)
traction_control = widgets.interactive(Select_Traction, Traction_Control=traction)
state_DD = widgets.interactive(Select_State, State=state)
cooling_DD = widgets.interactive(Select_Cooling, Cooling=cooling)
swingarm_DD = widgets.interactive(Select_Swingarm_Type, Swingarm_Type=swingarm)
colour_DD = widgets.interactive(Select_Colour, Colour=colour)
exhaust_DD = widgets.interactive(Select_Exhaust_Config, Exhaust_Config=exhaust)
front_brake_DD = widgets.interactive(Select_Front_Brake_Description, Front_Brake_Description=front_brake)
front_brake_cal_DD = widgets.interactive(Select_Front_Brake_Caliper, Front_Brake_Caliper=front_brake_caliper)
rear_brake_DD = widgets.interactive(Select_Rear_Brake_Description, Rear_Brake_Description=rear_brake)
rear_brake_cal_DD = widgets.interactive(Select_Rear_Brake_Caliper, Rear_Brake_Caliper=rear_brake_caliper)
frame_DD = widgets.interactive(Select_Frame_Material, Frame_Material=frame)
carburettor_DD = widgets.interactive(Select_Carburettor, Carburettor=carburettor)
front_suspension_DD = widgets.interactive(Select_Front_Suspension, Front_Suspension=front_suspension)
side_stand_DD = widgets.interactive(Select_Side_Stand, Side_Stand=side_stand)
cam_type_DD = widgets.interactive(Select_Cam_Type, Cam_Type=cam_type)


engine = widgets.IntText(value=750, description='Engine (cc):', disabled=False)
age = widgets.IntText(value=0, description='Age (years):', disabled=False)
kilometers = widgets.IntText(value=0, description='Kilometers:', disabled=False)
registration = widgets.IntText(value=6, description='Remaining_Registration (months):', disabled=False, style=style)
length = widgets.IntText(value=2000, description='Length (mm):', disabled=False, style=style)
trail = widgets.IntText(value=100, description='Trail (mm):', disabled=False, style=style)
front_brake = widgets.IntText(value=300, description='Front_Brake_Diameter (mm):', disabled=False, style=style)

# From layout
column1 = widgets.VBox([brand_DD, model_DD, state_DD, colour_DD, engine, age, kilometers, registration])
column2 = widgets.VBox([traction_control, cooling_DD, swingarm_DD, exhaust_DD,
                        front_brake_DD, rear_brake_DD, front_brake_cal_DD, rear_brake_cal_DD])
column3 = widgets.VBox([frame_DD, carburettor_DD, front_suspension_DD, side_stand_DD, cam_type_DD,
                       length,trail, front_brake])


button = widgets.Button(description='Predict Price',
                        disabled=False,
                        button_style='success',
                        tooltip='Predict the price of the bike with these features')

form = widgets.VBox([widgets.HBox([column1, column2, column3]),
            button])

out = widgets.Output()

display(form)

# Predict the price when button is clicked
button.on_click(predict_price)

```
