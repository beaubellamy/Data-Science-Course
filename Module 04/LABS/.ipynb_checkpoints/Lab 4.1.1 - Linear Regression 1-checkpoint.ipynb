#### 2.5  Predict Using Linear Regression

[Simple Linear Regression](https://en.wikipedia.org/wiki/Simple_linear_regression#Fitting_the_regression_line)

### $$\hat{y} = \beta_0 + \beta_1 x$$

> ### $$ \beta_1 = \frac{\sum_{i=1}^n (y_i - \bar{y} ) (x_i - \bar{x} )}{\sum_{i=1}^n (x_i - \bar{x})^2} $$

and

> ### $$ \beta_0 = \bar{y} - \beta_1\bar{x} $$

Given variables:
- $\bar{y}$ : the sample mean of observed values $Y$
- $\bar{x}$ : the sample mean of observed values $X$

##### 2.5 Define X, Y

- X = Predictor Variable 
- Y = Target Variable

```python
# ANSWER
X = data['sq__ft']
Y = data['price']
```

##### 2.5.A Calculate beta_1

 ### $$ \beta_1 = \frac{\sum_{i=1}^n (y_i - \bar{y} ) (x_i - \bar{x} )}{\sum_{i=1}^n (x_i - \bar{x})^2} $$

```python
# ANSWER
Xmean = X.mean()
Ymean = Y.mean()
numer = sum((Y - Ymean) * (X - Xmean))
denom = sum( (X - Xmean)**2 )

beta1 = numer/denom
```

##### 2.5.B Calculate beta_0

### $$ \beta_0 = \bar{y} - \beta_1\bar{x} $$

```python
# ANSWER
beta0 = Ymean - beta1 * Xmean
```

##### 2.4.C Find Yhat (Predicted Value Using Beta 0 and Beta 1)

### $$\hat{y} = \beta_0 + \beta_1 x$$

```python
# ANSWER
data['model_predict'] = beta0 + beta1 * data['sq__ft']
```

```python
plot_predictor(data, 'sq__ft', 'price', 'model_predict') 
```

##### 2.5.D Find Resudals

### $$ \text{residual}_i = y_i - \hat{y}_i$$

```python
# ANSWER
data['residuals'] = data['price'] - data['model_predict']
```

##### 2.5.D Find R-Square

The coefficient $R^2$ is defined as $$ R^2 = 1 - \frac{SS_{res}}{SS_{tot}} $$, where $$ SS_{res} $$ is the residual sum of squares $$ SS_{res} = \sum_{i=1}^n \left(y_i - \hat{y}_i\right)^2 $$ and $$ SS_{tot} $$ is the total sum of squares $$ SS_{tot} = \sum_{i=1}^n \left(y_i - \bar{y}\right)^2 $$

The best possible score is 1.0 and it can be negative (because the model can be arbitrarily worse).

**Write a function to calculate R- Square value.**

The total sum of squares:

### $$ SS_{tot} = \sum_{i=1}^n \left(y_i - \bar{y}\right)^2 $$

The residual sum of squares:

### $$ SS_{res} = \sum_{i=1}^n \left(y_i - \hat{y}_i\right)^2 $$

$R^2$:

### $$ R^2 = 1 - \frac{SS_{res}}{SS_{tot}} $$

```python
# ANSWER
def calculate_rsquared(target, predict):
    
    SST = sum((target - target.mean())**2)
    
    SSR = sum((target - predict)**2)
        
    return 1 - SSR/SST

print ('R squared: ',calculate_rsquared(data['price'], data['model_predict']))
```
